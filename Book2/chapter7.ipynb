```python
colors = ['k','g','r','b','c']
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define function
def f(x):
    return x**2  # Example function

# Define tangent line function
def approximate_tangent_line(x, slope, intercept):
    return slope * x + intercept

# Define x values and colors
x = np.linspace(-2, 5, 100)
y = f(x)
colors = ['red', 'blue', 'green', 'orange', 'purple']

plt.plot(x, y, label="f(x)")

for i in range(5):
    p2_delta = 0.0001
    x1 = i
    x2 = x1 + p2_delta
    y1 = f(x1)
    y2 = f(x2)

    # Approximate derivative
    approx_derivative = (y2 - y1) / (x2 - x1)
    b = y2 - approx_derivative * x2

    # Points to plot tangent line
    to_plot = [x1 - 0.9, x1, x1 + 0.9]

    # Scatter plot for the point of tangency
    plt.scatter(x1, y1, color=colors[i])

    # Plot tangent line and add label for legend
    plt.plot(
        to_plot,
        [approximate_tangent_line(x, approx_derivative, b) for x in to_plot],
        color=colors[i],
        label=f"Tangent at x={x1}: {approx_derivative:.4f}",
    )

    print(f"Approximate derivative for f(x) where x = {x1} is {approx_derivative}")

# Show legend
plt.legend()
plt.show()
```
